```python
#1st-Question
import numpy as np
arr=np.random.randint(1,101,size=(5,5)) #5*5 arr , filling it b/w 1 and 100
print(arr)

#a-part
middle=arr[2,2]
print('Middle Element', middle)

#b-part
for i in range(5):
  row_mean=np.mean(arr[i]) #arr[i] represents row as it is always for horizontal row
  print(f' Mean of row{i}: {row_mean}')

#c-part
overall_mean=np.mean(arr)
arr_abovemean=[]
for i in range(5):
  for j in range(5):
    if arr[i][j]>overall_mean:
      arr_abovemean.append(arr[i][j])

print(arr_abovemean)

#d-part
def spiral_order(matrix):
  result=[]
  top=0 #first row
  bottom=4 #last row
  left=0 #first column
  right=4 #last column

  while left<=right and top<=bottom:
    for i in range(left,right+1):
      result.append(matrix[top,i])
    top+=1

    for i in range(top,bottom+1):
      result.append(matrix[i,right])
    right-=1

    for i in range(right,left-1,-1):
      result.append(matrix[bottom,i])
    bottom-=1

    for i in range(bottom,top-1,-1):
      result.append(matrix[i,left])
    left+=1


  return result

print(spiral_order(arr))     #returning that array which we cretaed in the question in spiral format and then finally into list


```

```python
#2nd-Question
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/videogamesales.csv')

#a-part
df['global_sales']=df['NA_Sales']+df['EU_Sales']+df['JP_Sales']+df['Other_Sales']
df_sorted=df.sort_values('global_sales',ascending=False).reset_index(drop=True) #reset.index- I have used this for resetting row numbers
print(df_sorted)

#b-part
genre_sales=df.groupby('Genre')['global_sales'].sum()  #df.groupby('Genre')- groups all name with their genre and ['global_sales']-tells fromeach game, pick only their globalmsales no. and then sum it(.sum())
genre_sales.plot(kind='bar',color='blue',figsize=(10,6))
plt.title('Total Global Sales By Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales (millions)')
plt.tight_layout()
plt.show()

#c-part
gta=df[df['Name'].str.contains('Grand Theft Auto',na=False)] #na=False (if a game has no name(empty),just skip it), outer df is for selecting rows and inner df is for checking the conditions.
gta=gta.copy() #making a fresh copy to avoid any warnings.
gta['EU_JP_Sales']=gta['EU_Sales']+gta['JP_Sales']
gta_display=gta[['Name','Platform','Year','EU_JP_Sales']]
print(gta_display)

#d-part
gta_regions=gta[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
labels=['North America','Europe','Japan','Other']
sizes=gta_regions.values # .values-extracting just the numbers from gta_regions

plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title('Total GTA Sales by Region')
plt.tight_layout()
plt.show()
```
