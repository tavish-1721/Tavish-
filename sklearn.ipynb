```python
from sklearn.linear_model import LinearRegression

X=[[1],[2],[3],[4],[5]]
y=[40,50,65,76,90]

model=LinearRegression()

model.fit(X,y)

hours=float(input("Enter how many hours you studied="))

predicted_marks=model.predict([[hours]])

print(f"Based on your hours {hours}, you may score around {predicted_marks}")
```

```python
from sklearn.linear_model import LogisticRegression
X=[[1],[2],[3],[4],[5]]
y=[40,50,65,75,90]
model=LogisticRegression()
model.fit(X,y) #data passing
hours=float(input("Enter how many hours you studied="))
result=model.predict([[hours]])[0]
if result==1:
    print(f"you are likely to pass")
else:
    print(f"you are likely to fail")
    
    
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

X = [ [180,7], [200,7.5], [250,8], [300,8.5], [330,9], [360,9.5]]

# 0-Apple, 1-Orange
y = [0,0,0,1,1,1]

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X,y)

weight = float(input("Enter the weight in grams: "))
size = float(input("Enter the size in cm: "))

prediction = model.predict([[weight,size]])

if prediction == 0:
    print("This is likely an Apple")
else:
    print("This is likely an Orange")
    
        
    
    
```
