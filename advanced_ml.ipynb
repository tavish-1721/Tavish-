```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import matplotlib.pyplot as plt
import numpy as np

#load dataset
data= pd.read_csv("C:\\Users\\Tavish\\Downloads\\archive\\Housing.csv")

#input & output
X=data[['area','bedrooms','bathrooms','stories','parking']]
y=data['price']

model=LinearRegression()
model.fit(X,y)
predicted_price=model.predict(X)

#valid regression metrics
#evaluate
mae=mean_absolute_error(y,predicted_price)
mse=mean_squared_error(y,predicted_price)
rmse=np.sqrt(mse)
r2=r2_score(y,predicted_price)

#show results
print("Mean Squared Error: ",round(mse))
print("Mean Absolute Error: ",round(mae))
print("Root Mean Squared Error: ",round(rmse))
print("R^2 Score(Model Accuracy): ",round(r2,4)) #closer to 1,then it is better

#histogram
plt.figure(figsize=(10,6))
plt.hist(data["price"],bins=30,color='skyblue',edgecolor='black')
plt.title("Distribution of final price")
plt.xlabel("area")
plt.ylabel("price")
plt.grid(True)
plt.show()

#scatter plot + regression line
plt.figure(figsize=(8,8))

plt.scatter(y, predicted_price)

plt.xlabel("Area")
plt.ylabel("Predicted Price")

plt.title("Actual vs Predicted Prices")

plt.grid(True)
plt.show()

predicted_new_scores=model.predict([[9]])
print(f"Predicted Final Score for 9 Hours is {predicted_new_score}")
```
