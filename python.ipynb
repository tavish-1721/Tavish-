```python
#decorator-> adding another funcn inside funcn
def my_decorator(funcn):
  def wrapper():
    print("Something is happening before the function is called")
    funcn()
    print("Something is happening after the function is called")
  return wrapper

#say_hello = my_decorator(say_hello) is same as three lines below
@my_decorator
def say_hello():
  print("Hello")

say_hello()




```

```python
#generator
def count_down(num):
  while num>0:
    yield num # pauses here, returns value
    num-=1  #resumes from here next time

for number in count_down(5):
  print(number)
# 5  ← num=5, yield 5, then num=5-1=4
# 4  ← num=4, yield 4, then num=4-1=3
# 3  ← num=3, yield 3, then num=3-1=2
# 2  ← num=2, yield 2, then num=2-1=1
# 1  ← num=1, yield 1, then num=1-1=0
#    ← num=0, while stops!
```

```python
#constructor in OOPs
#__init__() #constructor
"""syntax
class Classname:
  def __init__(self,parameter1,parameter2):
    self.property1=parameter1
    self.property2=parameter2

    constructor types-default constructor(self),parametrised constructor(self,parameter1,parameter2),onstructor with default values(self,name='unknown',age=0)"""


class Car():
  def __init__(self,brand,color):
    self.brand=brand
    self.color=color

car1=Car('Tesla','Blue') #values auto assigned without actually calling that funcn
print(car1.brand,car1.color)

#polymrphism-same method for different classes ,different objects,different data types(One name with many forms)
class Bird():
    print("Birds make sounds")
class Crow(Bird):  #inheriting properties from bird class
  def sound(self):
    print("Crows say Caw Caw")
class Parrot(Bird):
  def sound(self):
    print("Parrot says squawk")

bird1=Crow()
bird2=Parrot()

bird1.sound()
bird2.sound()

  #method name is same but the output is different, Child class is overwriting Parent class.

#encapsulation-hiding internal details of objects and we can't modify it as the details are hidden.

class BankAccount:
  def __init__(self,account_number,balance):
    self.account_number=account_number #public info, anyone can access it
    self.__balance=balance #private info (by adding '__')-hidden from public

  def deposit(self,amount):
    self.__balance+=amount
    print(f'Deposited {amount}. New  balance is {self.__balance}')

  def get_balance(self):
    return self.__balance #controlled access, since __balance is pvt -can't access directly, use get_balance() to read it.

account= BankAccount('12345',5000)

account.deposit(2000)
print(account.get_balance())

#print(account.__balance) -private info can't be accessed


#Inheritance
class Animal:
  def speak(self):
    print("Animals make sound")

class Dog(Animal):  #inherited
     def bark(self):
      print('Dogs bark')

dog=Dog()
dog.speak() #using properties from parent
dog.bark()  #using own property

#Abstract class method-to hide complex details from user and get only those details which are necessary
from abc import ABC,abstractclassmethod  #abc = Abstract Base Class module,ABC = class to inherit from,abstractclassmethod = decorator for abstract methods
class Vehicle(ABC): # inheriting properties from abstract class
  @abstractclassmethod  #Forces ALL child classes to implement start() method
  def start(self):
    pass #no implementation, No code here, Child classes will decide HOW to implement start()

class Car(Vehicle):
  def start(self):
    print('Car starts with a key')

class Bike(Vehicle):
  def start(self):
    print('Bike starts with a button')

car=Car()
bike=Bike()

car.start()
bike.start()
```

ERRORS:-
1)compile time error-error in writing the code,syntax
2)run time error- like executing 2%0 will give this error when program is running.
3)logical error-unexpected outputs,like we want to add but given multiply sign.

errors in software- bug
eliminating these errors- debugging

runtime error can't be seen easily so in order to fix it we need to do exception handling.

exception-special type of runtime error (when program are running, warning sign is given, if catched then can be fixed and in this way we can save our program from getting crashed)

Like suppose tere are 100 lines of code and in that 100 lines, there are some 10-20 lines where errors may occur, so we will do exception handling in that 10-20 lines only , so that if error is there ,despite that program will run smoothly and at the end we will fix it.

```python
#Exception handling
#try-except(try block -error code (risky part of code) ,except block- if error occurs(what to do))
try:
  num=int(input('enter a number= '))
  result=10/num
  print(result)
except ZeroDivisionError:
  print('You cannot divide with 0')
except ValueError:
        print('You cannot divide with string')

#finally block- will always run despite of the exception raised or not
# try:
#   file=open("path")
#   content=file.read()
#   print(content)
# except FileNotFoundError:
#   print('File Not Found')

# finally:
#   file.close()
#   print('file operation completed')

#nested try-except block
try:
  num1=(int(input('enter number 1= ')))
  num2=(int(input('enter number 2= ')))
  try:
    result=num1/num2
    print(result)
  except ZeroDivisionError:
    print('You cannot divide with 0')
except ValueError:
  print('You cannot divide with string')

def check_password(password):
  if len(password)<8:
    raise Exception('Password is too short')
try:
  password=input('enter a password= ')
  check_password(password)
except Exception:
  print(Exception)
```
