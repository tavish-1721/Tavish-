```python
import torch

x = torch.tensor(2.0)
w = torch.tensor(3.0, requires_grad=True)
y_true = torch.tensor(10.0)
lr = 0.1

# --- before update ---
y_pred = w * x
loss = (y_pred - y_true) ** 2
loss.backward()

print("Before update")
print("w:       ", w.item())          # 3.0
print("loss:    ", loss.item())       # 16.0
print("gradient:", w.grad.item())     # -16.0
"""loss = (w*x - y_true)²
     = (w·2 - 10)²
∂loss/∂w = 2 · (w·x - y_true) · x
          = 2 · (3·2 - 10)   · 2
          = 2 · (6 - 10)     · 2
          = 2 · (−4)         · 2
          = −16"""

# --- update ---
with torch.no_grad(): #no graph recorded,stops pytorch gradient tracking,w updated safely,without this:w update would break the graph
    w -= lr * w.grad
w.grad.zero_() #making the gradient zero otherwise gradients add into the first

# --- recompute after update ---
y_pred = w * x
loss = (y_pred - y_true) ** 2

print("\nAfter update")
print("w:       ", w.item())          # 4.6
print("loss:    ", loss.item())       # 0.64  ← now correct
print("gradient:", w.grad)            # None (zeroed in w.gead_zero, no new backward yet)
```

```python
import torch
x= torch.tensor([1.,2.,3.,4.])
y_true=torch.tensor([2.,4.,6.,8.]) #correct values
w=torch.tensor(0.0,requires_grad=True)

lr=0.1
epochs=10 #no. of times AI have to learn

for epoch in range(epochs):
    #step 1 prediction
    y_pred=w*x
    #step  2 loss
    loss=((y_pred-y_true)**2).mean()
    loss.backward()

    with torch.no_grad():
        w-=lr*w.grad

    print(f"Epoch{epoch+1}:w={w.item():.4f},loss={loss.item():.4f}")
```

```python
#Neural Network,class
import torch
import torch.nn as nn

"""classMyModel(nn.Module):
def __init_(self):
super()._init_()"""
class MyModel(nn.Module): #neural network class is created, and this class is following PyTorch rules-numbers should be trained,learning allowed,data set can be loaded and saved and model can be optimized
  def __init__(self):   #what should be already there in model initially when created-learning no. ,layers,internal parts
    super()._init__() # without this nothing can be inherited from nn.module and model will break

    self.w=nn.Parameter(torch.tensor(0.0)) #self.w=nn.Parameter(torch.tensor(0.0)) replaces w=troch.tensor(0.0,requires_grad=True) automatically,nn.parameter tells PyTorch to track its gardient automatically.
    def forward(self,x): #defines the forward pass-what the model does with input(x),when we will call model(x),it will do w*x
     return self.w*x
```

```python
#Activation Functions,no activation means no intelligence
#There may be 5 layers but it will behave like a single layer ,model will become useless for a complex model because output=input*weight+bias is a linear formula
#model will not be able to learn curves,how to find shapes and also not be able to learn from data
#activation function is like a decison maker that decides which information shpould be passed forward and which info should be blocked.
#Linear-activation-linear-activation-output
#nn.Relu(Hidden Layer)-Rectified linear unit,-ve data becomes 0 and positive will be positive
#nn.Sigmoid(Binary Output) -used only in final output,Yes/No,0/1
#nn.Softmax(Multiclass output)-takes many numbers and convert them into one single number,or one single strongest other becomes week like [2,1,0]-[0.7,0.2,0.1],now 0.7 is the strongest
#input-linear-Relu-linear-Softmax/Sigmoid-output
```

```python
#optimiser is like a helper that automatically update all learning numbers(w1,w2,w3,..) using gradient like we do not have to use now w-=lr*w.grad
#all training steps will be there in gradient,decides how much to change each number by,clears the old gradient

#Optimisers type
#SGD-simple,perfect for basic learning,move towards its target slowly,lr=0.01
#torch.optim.SGD(model.parameter(),lr=0.01)
#Adam-auto adapt speed,perfect for faster learning,noisy data can be handled easily
#torch.optim.Adam(model.parameter(),lr=0.001)

#Step
#optimiser.zero_grad()-forward pass-loss caln -loss backward()-optimiser.step

import torch
import torch.nn as nn
import torch.optim as optim

class myModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.Linear=nn.Linear(1,1)
  def forward(self,x):
    return self.Linear(x)

x=torch.tensor([[1.],[2.],[3.],[4.]]) #samples-[1,4]
y=torch.tensor([[2.],[4.],[6.],[8.]]) #shape-[1,4]

model= myModel() #creates the model,initialises random weight
loss_fn=nn.MSELoss() #Mean squared error-measures how wrong predictions are

optimiser =optim.Adam(model.parameters(),lr=0.1) #adam(optimiser) will update all weights,parameters means using all learning numbers
for epoch in range(10): #loss get smaller with every next iteration
  optimiser.zero_grad() #clears gradients from previous epoch and avoids accumulation
  y_pred=model(x)
  loss=loss_fn(y_pred,y)#loss=mean((y_predict-y_true)^2)
  loss.backward() #computes gradient, backward pass-computes gradient of loss with respect to every learnable parameter(weight and bias),del loss/del weight,del loss/del bias,so loss is loss wrt weight +loss wrt to bias
  optimiser.step() #weights updated,adam uses the gradient to update weight and bias

  print(f"Epoch{epoch+1},Loss={loss.item():.4f}")
#after 10 epochs model.linear.weight will be close to 2.0 and module.linear.bias woill be close to 0.0,bcoz the relation is y=2x+0

torch.save(model.state_dict(),"linear_model.pth")
```
